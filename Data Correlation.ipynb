```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os, sys, re

file_path = r'C:\Users\shreyasb\Desktop\python'
Serial_flow = r'TGL_VIPR_TTR_PF_1.csv'
Parallel_flow = r'TGL_VIPR_TTR_Serial.csv'
data0 = pd.read_csv(file_path +'\\'+ Parallel_flow)

for col in data0.columns:
    if 'VIPRPINS' in col:
        data0[col].fillna('444|NO_DATA|-999', inplace = True)
        
for col in data0.columns:
    if 'CHUCKID' and 'DIERTD' in col:
        data0['CHUCKID'] = data0[col]

data0['UID'] = data0['LOT']+'_'+data0['X'].astype(str)+'_'+data0['Y'].astype(str)+'_'+data0['CHUCKID'].astype(str)
data0.head
```
